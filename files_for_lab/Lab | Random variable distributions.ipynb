```python
import pandas as pd
import seaborn as sns




data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')

data.head(5)
```

### Get the numerical variables from our dataset.

```python
num_data = data.select_dtypes(include='number')
```

```python
num_data.head(1)
```

### Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
for col in num_data.columns:
    sns.displot(num_data[col], kind='kde')
```

```python
import matplotlib.pyplot as plt

# create subplots for each variable
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(16, 8))

# plot distribution plot for each variable
for i, ax in enumerate(axes.flat):
    if i < len(num_data.columns):
        sns.histplot(num_data[num_data.columns[i]], kde=True, ax=ax)
        ax.set_title(num_data.columns[i])
    else:
        ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Months since policy inception is not far to be a normal distribution.
# Months since last claim also kind of look like a normal distribution

# Number of open complaints is exponential
```

### Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
import numpy as np

# log-transform numerical variables
num_data_log = np.log(num_data)

# plot distribution plot for each variable before and after the transformation
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(16, 8))

for i, ax in enumerate(axes.flat):
    if i < len(num_data.columns):
        sns.histplot(num_data_log[num_data.columns[i]], kde=True, ax=ax, color='orange', alpha=0.5, label='Log-Transformed')
        ax.set_title(num_data.columns[i])
        ax.legend()
    else:
        ax.axis('off')

plt.tight_layout()
plt.show()
```

```python


# plot distribution plot for each variable before and after the transformation
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(16, 8))

for i, ax in enumerate(axes.flat):
    if i < len(num_data.columns):
        sns.histplot(num_data[num_data.columns[i]], kde=True, ax=ax, color='blue', alpha=0.5, label='Original')
        sns.histplot(num_data_log[num_data.columns[i]], kde=True, ax=ax, color='orange', alpha=0.5, label='Log-Transformed')
        ax.set_title(num_data.columns[i])
        ax.legend()
    else:
        ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# The log-transform operation change the result on some variables. 

# On Customer Life Value we can see 3 pics that become more "normal"
# Month Since Policy Inception on the contrary was Normal distribution when it's now looks like exponential.
```
